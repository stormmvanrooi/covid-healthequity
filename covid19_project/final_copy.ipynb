```python
# Vaccination and Case Trends by Age

covid19_vacc_trends_url = "https://data.cdc.gov/resource/rh2h-3yt2.json"
response2 = requests.get(covid19_vacc_trends_url)
response2.json()
```

```python
vacc_data_df= pd.read_json(covid19_vacc_trends_url)
```

```python
location_groups = vacc_data_df.groupby('location')
location_groups.sum()

location_groups_clean = location_groups[['location','admin_dose_1_cumulative','series_complete_cumulative']]
location_groups_sum_df= location_groups_clean.sum()

location_groups_sum_df.sort_values(by=['series_complete_cumulative'], ascending=False)
```

```python
top_states_df = location_groups_sum_df.loc[['CA', 'TX', 'NY', 'FL', 'IL']]
bottom_states_df = location_groups_sum_df.loc[['SD', 'VT', 'AK', 'ND', 'WY']]
```

```python
top_states_df.plot(kind="bar", colormap='Accent')
plt.legend(loc="center", bbox_to_anchor=(0.5, 1.15), ncol=2)
x_axis = np.arange(len(top_states_df))
tick_locations = [value for value in x_axis ]
plt.xticks(tick_locations, ["California" , "Texas", "New York", "Florida", "Illinois"],rotation=25)
plt.xlabel("State")
plt.ylabel("Number of People Vaccinated")


plt.show()

bottom_states_df.plot(kind="bar",colormap='Accent')
plt.legend(loc="center", bbox_to_anchor=(0.5, 1.15), ncol=2)
tick_locations = [value for value in x_axis ]
matplotlib.pyplot.ylim([0, 1e8])
plt.xticks(tick_locations, ["South Dakota" , "Vermont", "Alaska", "North Dakota", "Wyoming"],rotation=25)
plt.xlabel("State")
plt.ylabel("Number of People Vaccinated")


plt.show()
```
